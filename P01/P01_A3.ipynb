### Aufgabe 3 git-Integration

In `work` bleibt Ihre Arbeit erhalten. Wir können aber keine Garantie übernehmen, dass das immer so ist, u.A. weil das JupyterHub-System bisher auf nur einem einzigen Server läuft.

Grundsätzlich können Sie natürlich Ihre Arbeit herunterladen und lokal speichern. Allerdings geht auch das nur auf der Dateiebene und nicht mit kompletten Ordnern (Sie können allerdings mehrere Dateien auswählen und mit `Download` aus dem Kontextmenu als `zip`-Datei herunterladen)

In der Vorlesung haben Sie git kennengelernt. Wir werden git zur Sicherung aber auch zur Abgabe von Hausarbeiten und zur Kontrolle der Anwesenheit bei Praktika nutzen.

#### Einrichten lokaler git-Repositories für Ihre Ordner `Praktikum` und `Hausarbeiten`
1. Wechseln Sie im Filebrowser in das Verzeichnis `Praktikum` (innerhalb Ihres `work`)
2. Legen Sie hier eine Datei `.gitignore` an. Dazu starten Sie ein *Terminal* aus dem *Launcher* und geben dort das Linux-Kommando 

`echo ".ipynb_checkpoints/" > .gitignore`

ein (Ausführen mit `Enter`)

3. Wechseln Sie im linken Panel auf "Git" (Klicken auf das Git-Symbol am linken Bildrand) und Klicken Sie auf "Initialize a Repository"
4. Sie sehen dann diese Darstellung




5. Bewegen Sie den Cursor auf "Untracked" und Klicken Sie auf das erscheinende "+". Sie verschieben damit die Dateien `.gitignore` und die Praktikums-Notebooks aus "Untracked" in "Staged (und nehmen damit die Dateien "unter  Versionsverwaltung") 

6. Schreiben Sie einen sinnvollen Eintrag in "Summary" (z.B. "initial commit") und klicken Sie auf "Commit". Damit erzeugen Sie den ersten "Schnappschuss" Ihre Praktikumsverzeichnisses. (Sie müssen Ihren Namen und Ihre E-Mail-Adresse angeben)

7. Wiederholen Sie diese Schritt 1-6 (Anlegen von `.gitignore`, Initialisieren und Commiten) in Ihrem `Hausarbeiten`-Verzeichnis. Ignorieren Sie die erscheinende Warnung

### Git-Routinearbeiten

Nach diesen einmaligen Einrichtungsarbeiten commiten/pushen Sie künftig jede signifikante Änderung in Ihren Ordnern mit den folgenden einfachen Schritten

#### Praktikum

1. Wechseln Sie auf das git-Panel
2. Checken Sie, ob neue Dateien unter `Untracked` in Versionsverwaltung genommen werden soll
3. Checken Sie Ihre Änderungen und *stagen* Sie die mit "+"
4. Committen Sie und pushen Sie (Sie können nach der Ersteinrichtung mit den Cloud-Buttons arbeiten)

#### Hausarbeiten

1. Wechseln Sie auf das git-Panel
2. Checken Sie, ob neue Dateien unter `Untracked` in Versionsverwaltung genommen werden soll
3. Checken Sie Ihre Änderungen und *stagen* Sie die mit "+"
4. Committen Sie und pushen Sie (Sie können nach der Ersteinrichtung mit den Cloud-Buttons arbeiten)
