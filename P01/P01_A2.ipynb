### Aufgabe 2 Einrichten des `work`-Ordners

Wir befinden uns noch in einem Unterordner zu `IT2`. Hier können Sie sich umschauen, auch Notebooks ausführen aber nicht speichern. Speichern können Sie ausschließlich unter `work`. Da Sie vermutlich im Laufe Ihres Studiums mehr als einen Kurs auf dem JupyterHub belegen werden, arbeiten Sie bitte in Unterverzeichnissen unter `work`.

1. Wechseln Sie im Filebrowser in das `home`-Verzeichnis (das Karteikartensymbol ganz links) und öffnen Sie mit Doppellklick `work`

 

2. Legen Sie hier einen neuen *Folder* an (Karteikartensymbol mit "+" oder rechte Maustaste) und benennen Sie ihn `IT2_2324`. Innerhalb dieses Ordners erzeugen Sie zwei weitere Unterordner `Praktikum` und `Hausarbeiten`



3. Unterhalb von `Praktikum` legen Sie für jeden Termin einen neuen Unterordner an - für den ersten Termin `P01` - und kopieren die Inhalte des Ordners `IT2/Praktikum/P01` hier hinein
>Es gibt (leider) keinen Befehle, um alle Inhalte eines Ordners zu kopieren. Sie können aber mehrere Dateien markieren, kopieren und an neuer Stelle einfügen

4. In `Hausarbeiten` legen Sie bitte eine neue Textdatei folgendem Inhalt an

```
Datum, Inhalt, Bemerkungen, Unterschriftscode
13.10.2023,Einführung,,
20.10.2023,Python-Grundlagen,,
```
und speichern diese als `IT2_1234567_Testate.csv` (mit Ihrer Matrikelnummer!)

>
**Die Schritte zum Kopieren der Inhalte aus `IT2` in Ihren `work`-Ordner führen Sie bitte zu Beginn jedes Praktikums aus!**
