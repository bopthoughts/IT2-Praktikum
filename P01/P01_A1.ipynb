### Grundelemente im Überblick

- Oben ist die Menuzeile. Praktisch alle Befehle haben hier einen Eintrag. 
- Ganz unten ist eine Statuszeile mit vielen nützlichen Informationen
- Der Bildschirm dazwischen ist horizontal in drei Abschnitte (Panels) unterteilt, wovon das mittlere immer sichtbar ist und die Panels links und rechts über Buttons am Rand ein- und ausgeklappt werden können. Im Startzustand ist links der File-Browser und das rechte Panel ist eingeklappt.
- Das mittlere Panel beinhaltet im Startzustand den *Launcher* mit dem z.B. neue Notebooks erstellt werden. 
- Wenn Sie einen Eintrag im Filebrowser anklickern, wird ein neues Fenster mit der jeweiligen Datei im mittleren Panel geöffnet.
(Über das "+" im Filebrowser können Sie jederzeit ein neues *Launcher*-Fenster öffnen, falls Sie es geschlossen haben)
- Viele (aber nicht alle) Menueinträge erreichen Sie auch über Buttons am Bildschirmrand und/oder über den Panels
- Viele (aber nicht alle) Menueinträge erreichen Sie auch über Tastenkombinationen. Welche es gibt, erfahren Sie über `Settings`->`Advanced Settings Editor` aus dem Hauptmenu. Dort können Sie auch Key-Bindings ändern. Machen Sie davon aber nur Gebrauch, wenn es unumgänglich ist.
